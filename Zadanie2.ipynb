```python
import requests
import json

API_KEY_VULNERS = '9OAGIH4UQAW1L0BUVT1QWQEYOJYT2M18H7BCYOKE8HO98XSE6ASRQN6G6Z9GIZT8'
VULNERS_API_URL = 'https://vulners.com/api/v3/search/lucene/'

def load_software():
    with open('my_programs.json') as f:
        return json.load(f)

def search_vulns(program, version):
    try:
        response = requests.get(VULNERS_API_URL, params={
            'apiKey': API_KEY_VULNERS,
            'query': f'"{program} {version}"',
            'size': 10
        })
        response.raise_for_status()
        return response.json().get('data', {}).get('search', [])
    except Exception as e:
        print(f"Ошибка для {program}: {str(e)}")
        return []

if __name__ == '__main__':
    software_list = load_software()
    
    for item in software_list:
        print(f"\nАнализ: {item['Program']} {item['Version']}")
        vulns = search_vulns(item['Program'], item['Version'])
        
        if vulns:
            print("Найдены уязвимости:")
            for vuln in vulns:
                cve = vuln['_source'].get('cvelist', [])
                print(f"  CVE: {', '.join(cve) if cve else 'Нет данных'}")
                print(f"  Эксплойт: {'Да' if vuln['_source'].get('exploit_available') else 'Нет'}")
        else:
            print("Уязвимости не найдены")
        print("-"*30)
```
